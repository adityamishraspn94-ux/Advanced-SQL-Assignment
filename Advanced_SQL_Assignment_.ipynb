### **Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?**

**Answer:**

A **Common Table Expression (CTE)** is a temporary named result set created using the `WITH` clause in SQL.

**Improvements in readability:**

* Breaks complex queries into logical steps
* Replaces nested subqueries with clean structure
* Improves understanding of query flow
* Makes debugging easier
* Enhances code maintainability

### **Q2. Why are some views updatable while others are read-only? Explain with an example.**

**Answer:**

Views behave differently based on their structure.

**Updatable Views:**

* Created from a single table
* No `JOIN`, `GROUP BY`, `DISTINCT`, or aggregate functions
* Direct mapping to base table rows

**Read-only Views:**

* Created using multiple tables (`JOIN`)
* Use aggregation functions (`SUM`, `COUNT`, etc.)
* Use grouping or calculations

**Explanation:**
Simple views allow data modification, while complex views with aggregation and joins do not support updates.

### **Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?**

**Answer:**

**Advantages of Stored Procedures:**

* Code reusability
* Faster execution (precompiled)
* Improved security (controlled access)
* Reduced network traffic
* Centralized business logic
* Easy maintenance and updates

### **Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.**

**Answer:**

**Purpose of Triggers:**

* Automatic execution on data events
* Enforce business rules
* Maintain data integrity
* Enable auditing and logging
* Ensure consistency

**Essential Use Case:**
Automatically recording changes in sensitive data (e.g., salary, price, stock) in an audit table whenever an update occurs.

### **Q5. Explain the need for data modelling and normalization when designing a database.**

**Answer:**

**Need for Data Modelling:**

* Defines database structure clearly
* Establishes relationships between tables
* Supports scalability and expansion
* Improves system design quality

**Need for Normalization:**

* Eliminates data redundancy
* Prevents data anomalies
* Improves data consistency
* Saves storage space
* Enhances data integrity
